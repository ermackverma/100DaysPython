```python
student_scores = {
    'Harry': 88,
    'Ron': 78,
    'Hermione': 95,
    'Draco': 75,
    'Neville': 60
}

# Loop through each key in the student_scores dictionary
for student in list(student_scores.keys()):

    # Get the value (student score) by using the key each time.
    score = student_scores[student]

    # Check what grade the score would get, then add it to student_scores
    if score >= 91:
        grade = 'Outstanding'
    elif score >= 81:
        grade = 'Exceeds Expectations'
    elif score >= 71:
        grade = 'Acceptable'
    else:
        grade = 'Fail'

    student_scores[student+'_grade'] = grade

print(student_scores)
```
